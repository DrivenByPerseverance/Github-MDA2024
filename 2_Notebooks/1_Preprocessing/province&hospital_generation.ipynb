#### Generate province variable for intervention dataset and generate hospital dataset

In this notebook, we use geopandas package to generate the province variable for intervention dataset. The Belgium province geo data is `1_Data/Belgium.provinces.WGS84.geojson` from https://github.com/mathiasleroy/Belgium-Geographic-Data/blob/master/dist/polygons/geojson/Belgium.provinces.WGS84.geojson .

```python
itv_ca = pd.read_parquet(
    '/Users/lye/Downloads/MDA/Github-MDA2024/1_Data/CLEANED/subset_cardiac.parquet',
    engine='pyarrow')

itv_ca.head()
```

```python
itv_ca.rename(columns={
    'latitude_intervention': 'lat_itv',
    'longitude_intervention': 'lon_itv',
    'province_intervention': 'province'
}, inplace=True) ## rename columns

itv_ca = itv_ca.dropna(subset=['lat_itv', 'lon_itv'])  ## drop rows with missing values
itv_ca.reset_index(drop=True, inplace=True)

itv_ca.head()
```

```python
import geopandas as gpd

geo_path = '/Users/lye/Downloads/MDA/Github-MDA2024/1_Data/Belgium.provinces.WGS84.geojson'
geo_be = gpd.read_file(geo_path)

geo_be.head(10)
```

```python
geo_be['NameDUT'].value_counts()
```

```python
from shapely import geometry as geo
from shapely.validation import explain_validity
# Assuming Belgian provinces are polygons in the geojson file
# We can check the in which province the intervention is located
point = geo.Point(itv_ca.loc[0, 'lon_itv'], itv_ca.loc[0, 'lat_itv'])
geo_be[geo_be.geometry.contains(point)]['NameDUT'].values[0]
```

```python
# Check if the geometries are valid

for i in range(len(geo_be)):
    if not geo_be.loc[i, 'geometry'].is_valid:
        print(explain_validity(geo_be.loc[i, 'geometry']))
        geo_be.loc[i, 'geometry'] = geo_be.loc[i, 'geometry'].buffer(0)
        print(geo_be.loc[i, 'geometry'].is_valid)
```

```python
def get_province(df, geo_df):
    province = []
    missing_province = 0
    for i in range(len(df)):
        point = geo.Point(df.loc[i, 'lon_itv'], df.loc[i, 'lat_itv'])
        contained = geo_df.loc[geo_df['geometry'].contains(point)]['NameDUT'].values
        if contained.size > 0:
            province.append(contained[0])
        else:
            province.append(None)
            missing_province += 1

    df['province'] = province
    print(f'{missing_province} coordinates are not located in any province')

    return df


itv_ca_with_province = get_province(itv_ca, geo_be)
itv_ca_with_province['province'].value_counts()
```

```python
itv_ca_with_province.info()
```

```python
itv_ca_with_province = itv_ca_with_province.dropna(subset=['province'])

itv_ca_with_province.to_csv(
    '/Users/lye/Downloads/MDA/Github-MDA2024/1_Data/CLEANED/subset_ca_with_province.csv',
    index=False)
```

```python
## Do the same for the whole intervention dataset

itv.rename(columns={
    'latitude_intervention': 'lat_itv',
    'longitude_intervention': 'lon_itv',
    'province_intervention': 'province'
}, inplace=True)  ## rename columns

itv = itv.dropna(subset=['lat_itv', 'lon_itv'])  ## drop rows with missing values
itv.reset_index(drop=True, inplace=True)


itv_with_province = get_province(itv, geo_be)
itv_with_province['province'].value_counts()
```

```python
itv_with_province = itv_with_province.dropna(subset=['province'])
itv_with_province.to_csv(
    '/Users/lye/Downloads/MDA/Github-MDA2024/1_Data/CLEANED/interventions_with_province.csv',
    index=False)
```
